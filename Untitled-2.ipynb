```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
print(df.head)
```

```python
print(df.shape)
print(df.dtypes)
print(df.isna().sum().sort_values(ascending=False))
```

```python
df = df.drop(columns = ["Cabin"])#dropping the column Cabin since it has many missing values

df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])#fill the missing value with the most frequent value 

df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
print(df.columns)
print(df.isna().sum().sort_values(ascending=False))
```

```python
df.groupby("Sex")["Survived"].mean()#survival by gender
```

```python
df.groupby("Sex")["Survived"].mean().plot(kind="bar")
plt.title("Average Survival rate by sex")
plt.ylabel("Survival rate")
plt.xlabel("Sex")
plt.show()
```

```python
df.groupby("Pclass")["Survived"].mean()
```

```python
df.groupby("Pclass")["Survived"].mean().plot(kind = "bar")
plt.ylabel("Survival rate")
plt.show()
```

```python
df["Sex"] = df["Sex"].map({"male":0, "female":1}) #label encoding
df[["Sex"]].head()
```

```python
df = pd.get_dummies(df, columns=["Embarked"],drop_first = True)
df.head()
```

```python
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1 #SibSp- siblings and spouse Parch- parents and children
df[["SibSp", "Parch", "FamilySize"]].head(10)
```

```python
df[["Sex"]].head()
```

```python
df.groupby("FamilySize")["Survived"].mean().plot(kind = "bar")
plt.title("Survival Rate by Family Members")
plt.xlabel("FamilySize")
plt.ylabel("Survival Rate")
plt.show()
```

```python
features = ['Pclass', 'Sex', 'Age', 'Fare', 'FamilySize', 'Embarked_S', 'Embarked_Q']
target = 'Survived'

X = df[features]
Y = df[target]

X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)

print(X_train.shape)#training set
print(X_test.shape)#testing set
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter = 1000)
model.fit(X_train, Y_train)
print("Model trained succesfully")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(Y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(Y_test, y_pred))
print("\nClassification Report:\n", classification_report(Y_test, y_pred))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

coef_df = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": model.coef_[0],
    "Odds_Ratio": np.exp(model.coef_[0])  # exp(coef) > 1 increases odds
}).sort_values("Coefficient", ascending=True)

display(coef_df)

plt.figure(figsize=(7,4))
plt.barh(coef_df["Feature"], coef_df["Coefficient"])
plt.title("Logistic Regression Coefficients (unscaled)")
plt.xlabel("Coefficient (log-odds)")
plt.ylabel("Feature")
plt.show()
```
